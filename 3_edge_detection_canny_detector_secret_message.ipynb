**Obtaining Secret Message from Image Using Canny Detector**

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from IPython.display import display
img = cv2.imread('frizzy.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_blur = cv2.GaussianBlur(img_gray, (3,3), 0)

img2 = cv2.imread('froomer.png')
img_gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
img_blur2 = cv2.GaussianBlur(img_gray2, (3,3), 0)

# Canny Edge Detection
edges = cv2.Canny(image=img_blur, threshold1=50, threshold2=100) # Canny Edge Detection
edges2 = cv2.Canny(image=img_blur2, threshold1=50, threshold2=100) # Canny Edge Detection

# Display Canny Edge Detection Image
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(15,10))

ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title("Frizzy")
ax2.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
ax2.set_title("Edges on Frizzy")
plt.savefig("result_frizzy.png")

fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2,figsize=(15,10))

ax1.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax1.set_title("Froomer")
ax2.imshow(cv2.cvtColor(edges2, cv2.COLOR_BGR2RGB))
ax2.set_title("Edges on Froomer")
plt.savefig("result_froomer.png")


fig, (ax1,ax2,ax3) = plt.subplots(nrows=1,ncols=3,figsize=(15,10))
ax1.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
ax1.set_title("Edges on Froomer")
ax2.imshow(cv2.cvtColor(edges2, cv2.COLOR_BGR2RGB))
ax2.set_title("Edges on Froomer")
ax3.imshow(cv2.cvtColor(edges & edges2, cv2.COLOR_BGR2RGB)) #Binary AND
ax3.set_title("Binary AND of Two Image")
plt.savefig("result.png")

fig, (ax1) = plt.subplots(nrows=1,ncols=1,figsize=(15,10))
ax1.imshow(cv2.cvtColor(edges & edges2, cv2.COLOR_BGR2RGB)) #Binary AND
ax1.set_title("Binary AND of Two Image")
plt.savefig("result2.png")
```
