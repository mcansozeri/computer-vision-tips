**Apple Sobel Filter To An Image**

```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('mcs.png')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
x = cv2.Sobel(gray, cv2.CV_32F, 1,0, ksize=3, scale=1)
y = cv2.Sobel(gray, cv2.CV_32F, 0,1, ksize=3, scale=1)
absx= cv2.convertScaleAbs(x)
absy = cv2.convertScaleAbs(y)
edge = cv2.addWeighted(absx, 0.5, absy, 0.5,0)
cv2_imshow(img)
cv2_imshow(edge)
```
